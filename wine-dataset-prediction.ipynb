```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
df = datasets.load_wine()
```

```python
df
```

Making a dataframe

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

## Independent Features and Dependent Features
   X is independent feature
   
   y is dependent feature

```python
x = dataset
y = df.target
```

```python
y
```

```python
x
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
      x, y, test_size=0.30, random_state=42)
```

```python
x_train
```

# Linear regression 

Do we need to standardize or normalize the data?
- In most cases we do standarization
for standardizing we have to use standard scalar

The standard score of a sample x is calculated as:
z = (x - u) / s

Standarizing the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
#cross validation
from sklearn.model_selection import cross_val_score
```

```python
#regression object 
regression = LinearRegression()
regression.fit(x_train ,y_train)
```

```python
mse = cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv= 10)
```

# For calculating the mean

Mean Squared error - it is the difference of the predicted value and the truth value . it should always be minimal.

```python
np.mean(mse)
```

```python
#prediction of test data
reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred - y_test,kind ='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(reg_pred,y_test)
```

```python
score
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge_regressor = Ridge()
```

```python
ridge_regressor
```

```python
parameters = {'alpha':[1,2,5,10,20,30,40,50,60,80,70,90]}
ridgecv = GridSearchCV(ridge_regressor, parameters , scoring = 'neg_mean_squared_error',cv = 5)
ridgecv.fit(x_train , y_train)
```

```python
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```

# Prediction

```python
ridge_pred = ridgecv.predict(x_test)
```

```python
import seaborn as sns
sns.displot(ridge_pred - y_test,kind ='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ridge_pred,y_test)
```

```python
score
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso()
```

```python
parameters = {'alpha':[1,2,5,10,20,30,40,50,60,80,70,90]}
lassocv = GridSearchCV(lasso, parameters , scoring = 'neg_mean_squared_error',cv = 5)
lassocv.fit(x_train , y_train)
```

```python
print(lassocv.best_params_)
print(lassocv.best_score_)
```

```python
lasso_pred = lassocv.predict(x_test)
```

```python
import seaborn as sns
sns.displot(lasso_pred - y_test,kind ='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(lasso_pred,y_test)
```

```python
score
```
